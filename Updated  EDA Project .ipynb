```python
merged_data = pd.merge(Transaction_data, Cab_Data, on ='Transaction ID', how ='left')
```

```python
new_data=pd.merge(merged_data, Customer_data)
```

```python
new_data_1=pd.merge(new_data, City_data)
```

```python
new_data_1["Gender"].value_counts()
```

```python
yellow_cab_prices = new_data_1[new_data_1["Company"] == "Yellow Cab"]["Price Charged"]
print("Price charged by Yellow Cab:")
print(yellow_cab_prices.describe())

pink_cab_prices = new_data_1[new_data_1["Company"] == "Pink Cab"]["Price Charged"]
print("Price charged by Pink Cab:")
print(pink_cab_prices.describe())
```

```python
# Similarly, we can also compare the cost of Trip between Yellow Cab and Pink Cab
yellow_cab_cost = new_data_1[new_data_1["Company"] == "Yellow Cab"]["Cost of Trip"]
print("Cost of Trip by Yellow Cab:")
print(yellow_cab_cost.describe())

pink_cab_cost = new_data_1[new_data_1["Company"] == "Pink Cab"]["Cost of Trip"]
print("Cost of Trip by Pink Cab:")
print(pink_cab_cost.describe())
```

```python
#Profit for Yellow and Pink Cabs

yellow_cab_profit = new_data_1[new_data_1["Company"] == "Yellow Cab"]["Price Charged"]-new_data_1[new_data_1["Company"] == "Yellow Cab"]["Cost of Trip"]
print("Profit of Yellow Cab:")
print(yellow_cab_profit.describe())

pink_cab_profit = new_data_1[new_data_1["Company"] == "Pink Cab"]["Price Charged"]-new_data_1[new_data_1["Company"] == "Pink Cab"]["Cost of Trip"]
print("Profit of Pink Cab:")
print(pink_cab_profit.describe())
```
